5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently
for each.


```python
my_list = [10, 20, 30]
print("Original list:", my_list)

my_list[1] = 200
print("Modified list:", my_list)

my_tuple = (10, 20, 30)
print("Original tuple:", my_tuple)

try:
    my_tuple[1] = 200
except TypeError as e:
    print("Error when modifying tuple:", e)
```

* Tuples are immutable, so attempting to change their elements results in a TypeError.

6. Write a function to demonstrate the behavior of mutable and immutable arguments.

▶  Explanation:
* Immutable types: int, float, str, tuple, etc. — cannot be changed in place.

* Mutable types: list, dict, set, etc. — can be changed in place.



```python
def modify_arguments(immutable_arg, mutable_arg):
    print("Inside function before modification:")
    print(f"immutable_arg: {immutable_arg}")
    print(f"mutable_arg: {mutable_arg}")

    # Try modifying both arguments
    immutable_arg += 10
    mutable_arg.append(4)

    print("Inside function after modification:")
    print(f"immutable_arg: {immutable_arg}")
    print(f"mutable_arg: {mutable_arg}")
```

```python
# Immutable argument (int)
num = 5

# Mutable argument (list)
lst = [1, 2, 3]

modify_arguments(num, lst)

print("\nOutside function after calling:")
print(f"num: {num}")       # unchanged
print(f"lst: {lst}")       # changed
```

7.  Write a program that performs basic arithmetic operations on two user-input numbers.

9.  Write a Python program to convert user input from string to integer, float, and boolean types.

```python

user_input = input("Enter a value: ")

try:
    int_value = int(user_input)
    print(f"Integer value: {int_value}")
except ValueError:
    print("Cannot convert to integer.")

try:
    float_value = float(user_input)
    print(f"Float value: {float_value}")
except ValueError:
    print("Cannot convert to float.")

true_values = {"true", "1", "yes", "on"}
false_values = {"false", "0", "no", "off"}

input_lower = user_input.strip().lower()

if input_lower in true_values:
    bool_value = True
elif input_lower in false_values:
    bool_value = False
else:
    bool_value = bool(user_input)

print(f"Boolean value: {bool_value}")
```

10.  Write code to demonstrate type casting with list elements.

```python
# Original list with string numbers
string_numbers = ["10", "20", "30", "40.5"]


int_numbers = [int(x) for x in string_numbers if x.isdigit()]
print("Casted to integers:", int_numbers)


float_numbers = [float(x) for x in string_numbers]
print("Casted to floats:", float_numbers)


str_numbers = [str(x) for x in float_numbers]
print("Casted back to strings:", str_numbers)


mixed_list = [1, "2", 3.5, "4.7", "hello", True]


str_list = [str(x) for x in mixed_list]
print("All elements as strings:", str_list)


int_list_safe = []
for item in mixed_list:
    try:
        int_list_safe.append(int(item))
    except ValueError:
        pass
print("Safely casted to integers:", int_list_safe)
```

14. Write a program to reverse a string using a while loop.

```python
original_string = input("Enter a string to reverse: ")

reversed_string = ""
index = len(original_string) - 1

while index >= 0:
    reversed_string += original_string[index]
    index -= 1

print("Reversed string:", reversed_string)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
num = int(input("Enter a non-negative integer: "))

factorial = 1
i = 1

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:

    while i <= num:
        factorial *= i
        i += 1

    print(f"The factorial of {num} is {factorial}")
```
